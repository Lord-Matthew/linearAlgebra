```
## START OF FUNCTION
import numpy as np
import matplotlib.pyplot as plt
def month_profit_trace(profit, reach, make_figs=True): 
    if (profit.shape == (4,)) and (reach.shape == (4,)): 
        e_week1 = np.array((reach[0], profit[0]))         ## Setup the weekly efficiency vectors 
        e_week2 = np.array((reach[1], profit[1]))
        e_week3 = np.array((reach[2], profit[2]))
        e_week4 = np.array((reach[3], profit[3]))
        
        e_week_total = np.add(e_week4, np.add(e_week3, np.add(e_week1,e_week2)))   
        e_week_stats = np.sqrt((e_week_total[0]**2)+(e_week_total[1]**2))
        alpha = 10**-6 
        theta = np.arctan(e_week_total[1]/(alpha + e_week_total[0]))     

        print(e_week_total)

        plt.figure(figsize=(16,5))
        plt.title('Bebang\'s Month Post Efficiency')
        plt.xlim(0,1.01*np.sum(reach))
        plt.ylim(-np.sum(np.abs(profit)),np.sum(np.abs(profit)))
        plt.xlabel('FB Post Reach Increment')
        plt.ylabel('Profit')
        plt.grid()
        n = 2
        
        plt.quiver(0,0, e_week1[0], e_week1[1], 
                   angles='xy', scale_units='xy',scale=1, color='pink', width=0.0025,
                   label='e_week1: {:.2f}'.format(np.linalg.norm(e_week1)))
        
        ## put necessary vector plotting code here

        plt.quiver(e_week1[0], e_week1[1], e_week2[0], e_week2[1], 
          angles='xy', scale_units='xy',scale=1, color='red', width=0.0025,
          label='e_week2: {:.2f}'.format(np.linalg.norm(e_week2)))
        plt.quiver(np.add(e_week1[0],e_week2[0]),np.add(e_week1[1],e_week2[1]),
          e_week3[0],e_week3[1],angles='xy',scale_units='xy',scale=1,color='yellow',width=0.0025, 
          label='e_week3: {:.2f}'.format(np.linalg.norm(e_week3)))
        x= np.add(e_week1[0],np.add(e_week2[0],e_week3[0]))
        y= np.add(e_week1[1],np.add(e_week2[1],e_week3[1]))
        plt.quiver(x,y,e_week4[0],e_week4[1], 
          angles='xy', scale_units='xy',scale=1,color='orange',width=0.0025,
          label= 'wk4: {:.2f}'.format(np.linalg.norm(e_week4)))
        plt.quiver(0,0,  e_week_total[0],  e_week_total[1],
          angles='xy', scale_units='xy',scale=1, color='green', width=0.005,
          label='Efficiency: {:.2f} @ {:.2f}'.format(e_week_stats, theta))

        plt.legend(loc='upper left')

        if make_figs:
            plt.savefig(f'LinAlg-Lab2-Bebang Post Eff-{int(e_week_total)}@{int(theta)}.png', dpi=300)

        plt.show()
    
    else:
        print('Please notify the incoder for profit and its reach is in value of 4') ## Make a more appropriate error statement. 


# # END OF FUNCTION 
```
